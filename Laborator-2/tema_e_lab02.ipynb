# Tema E - Lab 02 (Titanic)
Notebook cu analiza de baza a dataset-ului Titanic folosind pandas, seaborn si matplotlib.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
titanic.head()
```

## Statistici descriptive
Verificam dimensiunea, tipurile de date, sumarul numeric si valorile lipsa.

```python
print('Dimensiune:', titanic.shape)
print('\nTipuri de date:')
print(titanic.dtypes)
print('\nStatistici descriptive:')
print(titanic.describe(include='all'))
print('\nValori lipsa pe coloane:')
print(titanic.isnull().sum())
```

## Indicatori utili
Calculam rata de supravietuire globala, pe sex si pe clasa.

```python
rata_globala = titanic['survived'].mean()
rata_pe_sex = titanic.groupby('sex')['survived'].mean()
rata_pe_clasa = titanic.groupby('pclass')['survived'].mean()

print(f'Rata globala de supravietuire: {rata_globala:.2%}')
print('\nRata de supravietuire pe sex:')
print(rata_pe_sex)
print('\nRata de supravietuire pe clasa:')
print(rata_pe_clasa)
```

## Vizualizari
Creem doua grafice: distributia pasagerilor pe clasa si comparatia ratei de supravietuire pe sex.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.countplot(data=titanic, x='pclass', hue='sex', ax=axes[0])
axes[0].set_title('Numar pasageri pe clasa si sex')
axes[0].set_xlabel('Pclass')
axes[0].set_ylabel('Numar pasageri')

sns.barplot(data=titanic, x='sex', y='survived', errorbar='ci', ax=axes[1])
axes[1].set_title('Rata medie de supravietuire pe sex')
axes[1].set_xlabel('Sex')
axes[1].set_ylabel('Rata supravietuire')

plt.tight_layout()
plt.show()
```

## Concluzie
Datele arata diferente clare de supravietuire intre categoriile de pasageri.
Femeile au avut, in medie, o rata de supravietuire mai mare decat barbatii.
Clasa biletului influenteaza vizibil sansele de supravietuire, clasa 1 avand valori mai bune decat clasele inferioare.
Exista si valori lipsa in anumite coloane, ceea ce poate necesita preprocesare suplimentara in analize avansate.
